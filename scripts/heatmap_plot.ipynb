```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d
```

```python
def tunning(res_scale):
    num_rows, num_cols = 4, 8
    # res_scale = 10

    x, y = np.linspace(0, num_rows - 1, num_rows), np.linspace(0, num_cols - 1, num_cols)
    xx, yy = np.linspace(0, num_rows - 1, num_rows * res_scale), np.linspace(0, num_cols - 1, num_cols * res_scale)
    
    for k in ['linear', 'cubic']:
        f = interp2d(x, y, data, kind=k)
        highres_data = f(xx, yy).T

        sns.heatmap(highres_data, cmap = 'coolwarm')
        print(f'type: {k}\nscale: {res_scale}')
        plt.show()
```

```python
lowres_data = np.reshape(data, (4, 8))
sns.heatmap(lowres_data, cmap = 'coolwarm' )
plt.show()
```

```python
for scale in [2, 4, 10, 25]:
    tunning(scale)
```
